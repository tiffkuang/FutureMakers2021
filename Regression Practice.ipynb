```python

# -*- coding: utf-8 -*-
"""
Created on Wed Jul 15 00:20:02 2020

@author: Shoum
"""

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error
# Importing the dataset
names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
dataset = pd.read_csv("housing.csv", delim_whitespace=True, names=names)

# Divide the data into X and y
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values


# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, :] = sc.fit_transform(X_train)
X_val[:, :] = sc.transform(X_val)
#print(X_train)
#print(X_val)


# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
## create linear regression object
lin_reg = LinearRegression()
##train the model using train sets
lin_reg.fit(X_train, y_train)
## making predictions using train sets
lin_reg_pred = lin_reg.predict(X_train)
loss_linear_regression = mean_squared_error(y_train, lin_reg_pred)
print (loss_linear_regression)
lin_reg_pred = lin_reg.predict(X_val)
loss_linear_regression = mean_squared_error(y_val, lin_reg_pred)
print (loss_linear_regression)


# Training Support Vector Machine
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_train)
loss_svr = mean_squared_error(y_train, svr_pred)
print(loss_svr)
svr_pred = svr.predict(X_val)
loss_svr = mean_squared_error(y_val, svr_pred)
print(loss_svr)

# Training the Decision Tree Regression model on the whole dataset
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
tree_pred = tree.predict(X_train)
loss_tree = mean_squared_error(y_train, tree_pred)
print(loss_tree)
tree_pred = tree.predict(X_val)
loss_tree = mean_squared_error(y_val, tree_pred)
print(loss_tree)

# Traning Random Forest
from sklearn.ensemble import RandomForestRegressor
ensemble = RandomForestRegressor()
ensemble.fit(X_train, y_train)
ensemble_pred = ensemble.predict(X_train)
loss_random_forest = mean_squared_error(y_train, ensemble_pred)
print(loss_random_forest)
ensemble_pred = ensemble.predict(X_val)
loss_random_forest = mean_squared_error(y_val, ensemble_pred)
print(loss_random_forest)
```
